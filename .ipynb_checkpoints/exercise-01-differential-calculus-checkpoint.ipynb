```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# function defintion
def f(x):
    return x**3 - 2*x**2 + 1

# derivative by definition of limit
def f_derivative_by_limit(x, delta=0.0001):
    return (f(x + delta) - f(x)) / delta

# derivative by rules
def f_derivative_by_rules(x):
    return 3*x**2 - 4*x
```

```python
# create plots for the function and its derivative
def plot_functions(min_x=-2, max_x=3, delta=0.0001):
    x = np.linspace(min_x, max_x, 500)
    
    fig = plt.figure(figsize=(12,5))
    
    # left graph with f(x)
    ax = fig.add_subplot(121)
    y = f(x)
    ax.plot(x, y)
    ax.set_title("y=f(x)")
    
    # right graph with f'(x)
    ax = fig.add_subplot(122)
    y_by_limit = f_derivative_by_limit(x, delta)
    y_by_rules = f_derivative_by_rules(x)
    ax.plot(x, y_by_limit, c="r", alpha=0.5, label="rate")
    ax.plot(x, y_by_rules, c="b", alpha=0.5, label="rule")
    ax.set_title("y=f'(x)")
    ax.legend()
    plt.show()
```

```python
plot_functions()
```
